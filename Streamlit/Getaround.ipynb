```python
#Si delta = 0 A t-on plus de retard? 
df['time_delta=0']=df['time_delta_with_previous_rental_in_minutes']<1
df
```

```python
#Moyenne du retard par rapport à la course précédente si delta =0. Il y a un peu moins de retard si le delta est petit 
df.groupby(df['time_delta=0'])["delay_at_checkout_in_minutes"].mean()
```

```python
#On fait un merge... /On est sur une petite partie du dataset car previous rental id complété que sur 8% du dataset
df7 = pd.merge(df, df, how='inner', left_on = 'previous_ended_rental_id', right_on = 'rental_id')
df7
```

```python
fig = px.pie(df7['state_x'], names='state_x', title="Nombre d'annulations en proportion sur le dataset réduit")
fig.show()
```

```python
#Calcul du rapport entre le temps libre laissé entre la course précédente et la future course et le retard au check
df7['temps_libre']=df7['time_delta_with_previous_rental_in_minutes_x']-df7['delay_at_checkout_in_minutes_y']
```

```python
#Ajout de la colonne delta insufisant si le temps libre < au retard
df7['delta_insufisant']=df7['temps_libre']<0
```

```python
df7['delta_insufisant'].value_counts()
```

```python
fig = px.bar(df7['delta_insufisant'], x="delta_insufisant",color=df7["delta_insufisant"])
fig.show()
```

```python
fig = px.pie(df7['delta_insufisant'], names='delta_insufisant', title="")
fig.show()
```

```python
fig = px.histogram(df7['delta_insufisant'], x="delta_insufisant", color=df7["state_x"],barnorm='percent',text_auto=True,title="Proportion de temps libre insuffisant en fonction du state")
fig.show()
```

```python
insuf=df7.groupby(df7['delta_insufisant'])["delay_at_checkout_in_minutes_x"].mean()
insuf
```

```python
print("moyenne retard time delta  insuffisant:",round(insuf[0]),"min")
print("moyenne retard time delta  suffisant:",round(insuf[1]),"min")
```
